```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
X=df.drop('species',axis=1)
y=df['species']
```

#STACKING


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
base_learners=[
    ('lr', LogisticRegression(max_iter=1000)),
    ('dt', DecisionTreeClassifier()),
    ('svm', SVC(probability=True))
]
```

```python
meta_learner= LogisticRegression()
```

```python
# Stacking classifier
stack_model = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5
)
```

```python
#now train the model
stack_model.fit(X_train, y_train)
```

```python
stack_model.score(X_test, y_test)
```

#BaGGING : RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(
    n_estimators=100, #num of tress
    max_depth=None, #let te trees grow fully
    max_samples=0.8, # % of data to sample for each model
    max_features='sqrt', #no of features to consider randomly , square root of all feature
    random_state=10
)
```

```python
rf_model.fit(X_train,y_train)
```

```python
y_pred=rf_model.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

##xg boost model

```python
xg_model=XGBClassifier(
    n_estimator=100,
    learning_rate=0.1,
    max_depth=3,
    use_lable_encoder=False,
    eval_metric='logloss',
    random_state=41,
)
```

```python
from sklearn.preprocessing import LabelEncoder

# encode the target variable
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
xg_model.fit(X_train,y_train_encoded)
```

```python
y_pred=xg_model.predict(X_test)
```

```python
accuracy=accuracy_score(y_test_encoded,y_pred)
accuracy
```
